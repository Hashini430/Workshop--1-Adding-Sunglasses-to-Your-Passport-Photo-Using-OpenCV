```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
faceImage = cv2.imread('hash.jpg')

plt.imshow(faceImage[:,:,::-1])
plt.title("Face")
plt.show()
```

```python
glassJPG = cv2.imread('GLASS.JPG')

plt.imshow(glassJPG[:,:,::-1])
plt.title("Glass")
plt.show()
```

```python
glassJPG = cv2.resize(glassJPG,(470,95))
```

```python
glassBGR = glassJPG[:,:,0:3]
```

```python
glassGray = cv2.cvtColor(
    glassBGR,
    cv2.COLOR_BGR2GRAY
)
```

```python
_, glassMask1 = cv2.threshold(
    glassGray,
    240,
    255,
    cv2.THRESH_BINARY_INV
)
```

```python
plt.figure(figsize=[15,15])

plt.subplot(121)
plt.imshow(glassBGR[:,:,::-1])
plt.title('Glass Color')

plt.subplot(122)
plt.imshow(glassMask1,cmap='gray')
plt.title('Glass Mask')

plt.show()
```

```python
glassMask = cv2.merge(
    (glassMask1,glassMask1,glassMask1)
)

glassMask = np.uint8(glassMask/255)
```

```python
faceWithGlasses = faceImage.copy()
```

```python
eyeROI = faceWithGlasses[
    470:565,
    225:695
]
```

```python
maskedEye = cv2.multiply(
    eyeROI,
    (1-glassMask)
)

maskedGlass = cv2.multiply(
    glassBGR,
    glassMask
)
```

```python
eyeFinal = cv2.add(
    maskedEye,
    maskedGlass
)
```

```python
faceWithGlasses[
    470:565,
    225:695
] = eyeFinal
```

```python
plt.figure(figsize=[15,15])

plt.subplot(121)
plt.imshow(faceImage[:,:,::-1])
plt.title("Original")

plt.subplot(122)
plt.imshow(faceWithGlasses[:,:,::-1])
plt.title("With Glass")

plt.axis('off')
plt.show()
```
